60で構築したデータベースを用い，活動場所が「Japan」となっているアーティスト数を求めよ

http://gushwell.ldblog.jp/archives/52523792.html

```python
from redis import Redis
```

```python
r = Redis( )
```

```python
# 繰り返しスキャン
for i, key in enumerate(r.scan_iter()):
    
    #　HashesのValue取得
    for area in r.hvals(key):
        if area == b'Japan':
            print(key.decode())
```
